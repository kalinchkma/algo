# Polynomial Curve Fitting

### Data

- X label data x = 1 ..... N [N = 10]
- Y label data y = sin(2 * pi * x)

```python
n = 100
x = np.linspace(0, 1, n)
y = np.sin(2*np.pi*x)

# adding noise to y data
mean = 0 # mean of gaussian noise
std_dev = 0.1
noise = np.random.normal(mean, std_dev, y.shape)

pred_y = y + noise


x, y
```

```python


plt.plot(x, y, color="green")
plt.scatter(x, pred_y,)
plt.xlabel("Actual values")
plt.ylabel("sin values")
plt.title("sin wave scatter graph")

plt.show()
```
